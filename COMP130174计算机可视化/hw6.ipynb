```python
import numpy as np
import pandas as pd
import plotnine
plotnine .options .figure_size=(7.5, 4.2)
from plotnine import *
from plotnine import data
from pandas import DataFrame
```

```python
diamonds = data.diamonds
diamonds["nclarity"] = diamonds["clarity"]
clarity_nums = {
"nclarity" : {"I1" : 8,
"SI1": 7,
"SI2": 6,
"VS1": 5,
"VS2": 4,
"VVS1": 3,
"VVS2": 2,
"IF" : 1}
}
diamonds .replace(clarity_nums, inplace=True)
diamonds
```

```python
#取出价格在0-10000的
sdiamond = diamonds[diamonds["price"] < 10000]
sdiamond = sdiamond.sample(500)
```

```python
h6 = ggplot(sdiamond, aes(x="price", y="carat",size='clarity')) + geom_point()
```

```python
h6 = (ggplot(sdiamond, aes(x="price", y="carat",size = 'clarity',colour="cut")) + geom_point(shape = 'o') 
    + scale_x_continuous(breaks=np .arange(0, 10000, 1000))
    + scale_y_continuous(limits= (0, 2))
    + ggtitle("Carat of diamonds by price")
    + xlab("Diamond Price")
    + ylab("Carat")
    + theme_bw()
    + scale_colour_brewer(type="seq", palette="PuRd", name="Cut")
    )

h6

ggsave(h6,"p1.jpg",dpi=300)
```

```python
import matplotlib.font_manager as fm
fpath = "DecimaMonoPro.otf"
# Create font objects
title_text = fm .FontProperties(fname=fpath)
body_text = fm .FontProperties(fname=fpath)
# Alter size and weight of font objects
title_text .set_size(18)
title_text .set_weight("bold")
body_text.set_size(12)
```

```python
h6 = (ggplot(sdiamond, aes(x="price", y="carat",size = 'clarity',colour="cut")) + geom_point(shape = 'o') 
    + scale_x_continuous(breaks=np .arange(0, 10000, 1000))
    + scale_y_continuous(limits= (0, 2))
    + ggtitle("Carat of diamonds by price")
    + xlab("Diamond Price")
    + ylab("Carat")
    + theme_bw()
    + scale_colour_manual(values= ["gold", "limegreen", "dodgerblue","darkorchid", "red"])
    +theme(
        axis_line_x=element_line(size=2, colour="black"),
        axis_line_y=element_line(size=2, colour="black"),
        panel_grid_major=element_blank(),
        panel_grid_minor=element_blank(),
        panel_border=element_blank(),
        panel_background=element_blank(),
        plot_title=element_text(fontproperties=title_text),
        text=element_text(fontproperties=body_text),
        axis_text_x=element_text(colour="black"),
        axis_text_y=element_text(colour="black"),
    )
    )

h6
```
