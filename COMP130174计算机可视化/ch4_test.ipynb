```python
import numpy as np
import pandas as pd
import plotnine
plotnine .options .figure_size=(7.5, 4.2)
from plotnine import *
from pandas import DataFrame
```

```python
copper = pd.read_csv("copper.csv")
copper
```

```python
p4 = ggplot(copper, aes("year", "percentage", fill="product"))
p4
```

```python
p4 = (
ggplot(copper, aes("year", "percentage", fill="product"))
+ geom_col()
)
p4
```

```python
p4 = (ggplot(copper, aes("year", "percentage", fill="product"))
+ geom_col(position=position_stack(reverse=True))
)
p4
```

```python
p4 = (
ggplot(copper, aes("year", "percentage", fill="product"))
+ geom_col(position=position_stack(reverse=True))
+ geom_text(copper, aes(label="percentage"), size=9)
)
p4
```

```python
copper["pos"] = (copper[["year", "percentage"]] .groupby("year") .\
transform(pd .Series .cumsum))
copper["pos"] = copper["pos"] - 0.5 * copper["percentage"]
p4 = (
ggplot(copper, aes("year", "percentage", fill="product"))
+ geom_col(position=position_stack(reverse=True))
+ geom_text(aes(y="pos", label="percentage"), size=9,
format_string="{}%")
)
p4
```

```python
prod_name = {
"product" : {"copper" : "Copper",
"others" : "Pulp wood, Fruit, Salmon & Others"}
}
copper.replace(prod_name, inplace=True)
p4 = (
ggplot(copper, aes("year", "percentage", fill="product"))
+ geom_col(position=position_stack(reverse=True))
+ geom_text(aes(y="pos", label="percentage"), size=9,
format_string="{}%")
)
p4
```

```python
p4 = (
ggplot(copper, aes("year", "percentage", fill="product"))
+ geom_col(position=position_stack(reverse=True))
+ geom_text(aes(y="pos", label="percentage"), size=9,
format_string="{}%")
+ scale_x_continuous(breaks=np .arange(2006, 2015, 1))
)
p4
```

```python
p4=(
ggplot(copper, aes("year", "percentage", fill="product"))
+ geom_col(position=position_stack(reverse=True))
+ geom_text(aes(y="pos", label="percentage"), size=9,
format_string="{}%")
+ scale_x_continuous(breaks=np .arange(2006, 2015, 1))
+ ggtitle("Composition of Exports to China ($)")
+ xlab("Year")
+ ylab("USD million")
)
p4
```
